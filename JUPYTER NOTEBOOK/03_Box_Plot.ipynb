### use a pallet in data

### show means in the upper data

```python
sns.boxplot(x= "survived",
            y= "age",
            showmeans=True,
            data=kashti)
```

### Change the meanprops.

```python
import matplotlib.pyplot as plt
sns.boxplot(x= "survived",
            y= "age",
            showmeans=True,
            meanprops={"marker":"*",
                       "markersize": "12",
                       "markeredgecolor": "red"},
            data=kashti)
plt.xlabel("How many survived", size=10),
plt.ylabel("Age(Years)",size=10),
plt.title("Box plot of litny Doobe or Kitny Bach Gaye",size=14,weight='bold')
```
